```python
%matplotlib inline
import numpy as np
import pandas as pd

# Visualization
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
```

```python
# Select which data to load
filename = "cQuyen.TXT"

# Import number of channels
somerows = pd.read_csv(filename, skiprows=32, nrows = 1, sep='\t')
numchan = int(somerows.values[0][0][-1])

# Method to make labels automatically
def mklbl(prefix, n):
    return ["%s%s" %(prefix, i+1) for i in range(n)]

columnlbl = pd.MultiIndex.from_product([mklbl('Channel ', numchan),
                                       ['oxyHb', 'deoHb', 'totHb']])

# Import data array
data = pd.read_csv(filename, skiprows=35, sep='\t')
data1 = data.as_matrix()

# Create data frame
df = pd.DataFrame(data1[:, 4:], columns=columnlbl)
```

```python
from scipy import signal
import numpy as np

ts = 0.055
Fs = 1/ts
high_val = 0.01
low_val = 0.5

# Bandpass filter
cutfreq = np.array([high_val, low_val])
b, a = signal.ellip(4, 0.1, 40, cutfreq*2/Fs, btype='bandpass')
H, w = signal.freqz(b, a, 512)
```

```python
# Process a single time series
sig = df.T.loc['Channel 1', 'oxyHb']
sig = sig.as_matrix()
outsig = signal.filtfilt(b, a, sig)
```

```python
# Plot it
plt.plot(sig, 'k--', outsig, 'r-')
```

### Plot a single channel

```python
ch1 = df.T.loc['Channel 1']
ch1.T.plot(y=['oxyHb', 'deoHb', 'totHb'])
```

### Plot the whole data frame wrt Hb types

```python
def plot_Hb(mydf, hbtype):
    # Swap level (for plotting wrt Hb types)
    df_swap = mydf.swaplevel(0,1, axis=0)
    df_Hbtype = df_swap.loc[hbtype]
    df_Hbtype.T.plot(y=[i for i, data in df_Hbtype.iterrows()], ax=plt.gca())

plt.subplot(211), plot_Hb(df.T, 'deoHb')
```

## Bandpass filter
We will try to accomplish the task via 2 approaches
### Using apply()

```python
# Utilize apply
ch1filt = ch1.T
#filt = lambda x: x.mean()
filt = lambda x: signal.filtfilt(b, a, x)
ch1filt = ch1filt.apply(filt);
ch1filt.head()
```

```python
ch1filt.plot(y=['oxyHb', 'deoHb', 'totHb'])
```

```python
dffilt_byapply = df.apply(filt);
plt.subplot(211), plt.gca, plot_Hb(dffilt_byapply.T, 'oxyHb')
plt.subplot(212), plot_Hb(df.T, 'oxyHb')
```

### Defined function

```python
df.T.head()
```

```python
def filtfilt_df(mydf):
    dfout = mydf.copy()
    for i, data in mydf.iterrows():
        dfout.loc[i] = signal.filtfilt(b, a, data)
    return dfout

# apply in 1 channel
ch1filt = filtfilt_df(ch1)
ch1filt.T.plot(y=['oxyHb', 'deoHb', 'totHb'])
```

We acquired the same result. However, the defined function might be more effective in a broader functionality

```python
# apply in the whole data frame
dffilt_bydefinedfunct = filtfilt_df(df.T)

# Visualization
plt.subplot(211), plot_Hb(dffilt_bydefinedfunct, 'oxyHb')
plt.subplot(212), plot_Hb(df.T, 'oxyHb')
```

We obtained the same result using 2 different approaches. Hence, they are interchanable.
From now on, I will use defined function (just a arbitrary choice)

## Smoothing

```python
import sigpro as sp

def smooth_df(mydf, window_len=31):
    dfout = mydf.copy();
    for i, data in mydf.iterrows():
        dfout.loc[i] = sp.smooth(data, window_len)
    return dfout

dfbp_mva_sm = smooth_df(dfbp_mva, 511)
plt.subplot(211), plot_Hb(dfbp_mva_sm, 'oxyHb')
plt.subplot(212), plot_Hb(dfbp_mva, 'oxyHb')
```
