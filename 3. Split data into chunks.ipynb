### Make MultiIndex Dataframe based on Trials

```python
# Method to make labels automatically
def mklbl(prefix, n):
    return ["%s%s" %(prefix, i+1) for i in range(n)]

numtrial = 5
numchan = 7

len_task = task.shape[0]/numtrial

# Create trial-based data frame
miTrial = pd.MultiIndex.from_product([mklbl('Trial ', numtrial),
                                      [x for x in range(len_task)]])

miChHb = pd.MultiIndex.from_product([mklbl('Channel ', numchan),
                                       ['oxyHb', 'deoHb', 'totHb']])

dfmi = pd.DataFrame(task, index = miTrial, columns = miChHb)
dfmi
```

```python
# Sort Hb types in order to perfome slicing
dfmi_Hbsorted = dfmi.T.sort_index().sort_index(axis=1) # EXTREMELY IMPORTANT!!!!
dfmi_Hbsorted.head()
```

Here we encounter a big problem. The df.apply() approach allows us to only apply shifting on channel 1 but skip the others.
We have to use defined function approach here.
It's different from what we saw in Signal Processing where the 2 approaches return the same result.
The difference was caused by the multiindex, in this dataframe, both index and columns are multiindex.

Now we create a function that perfome both normalization and shifting to all time series (split by trials and channel-Hb types)

```python
# Normalizing in all trials
def nmlShiftAll(df):
    nml = lambda x: (x - x.mean()) / x.std() #normalization
    shift = lambda x: x - x[0]
    dfout = df.copy()
    idx = pd.IndexSlice
    for i in df.columns.levels[0]:
        a = df.loc[idx[:,:], idx[i, :]]
        a = a.apply(nml, axis=1)
        dfout.loc[idx[:,:], idx[i, :]] = a.apply(shift, axis=1)
        print i
    return dfout

dfmi_nmlShift = nmlShiftAll(dfmi_Hbsorted) # the function is quite slow
chunk = dfmi_nmlShift.loc[idx['Channel 1',:], idx['Trial 5', :]]
chunk
```

### Split data according to workload levels

```python
# paradigm: low-high-low-high-low
lvl1_trials = ['Trial 1', 'Trial 3', 'Trial 5']
lvl1 = dfmi_nmlShift.loc[idx[:, :], 
                        idx[lvl1_trials, :]]
lvl2_trials = ['Trial 2', 'Trial 3']
lvl2 = dfmi_nmlShift.loc[idx[:, :], 
                        idx[lvl2_trials, :]]

lvl1.head()
lvl1.columns.levels[0]
```

```python
a = lvl1.loc[idx['Channel 1', 'oxyHb'], idx[:, :]]
type(a)
```

```python
# reconstructing
def calc_mean(series_Hb, lvl_trials):
    data = series_Hb.values
    numtrial = len(lvl_trials)
    data = data
    data = np.reshape(data, (numtrial, len_task))
    #print data
    #print micolumns
    df = pd.DataFrame(data.T, index = [x for x in range(len_task)], columns = lvl_trials)
    return df

a2 = calc_mean(a, lvl1_trials)
a2.head()
```

```python
a2.plot()
```

```python
#func = lambda x: np.std(x)/np.sqrt(x.shape[1])
deno = np.sqrt(a2.shape[1])
m1 = a2.apply(np.sum, axis=1)
s1 = a2.apply(np.std, axis=1) / deno
```

```python
plt.plot(m1.index, m1, 'b')
plt.fill_between(s1.index, m1-2*s1, m1+2*s1, color='b', alpha=0.2)
```

### Plot the other mental worload level

```python
b = lvl2.loc[idx['Channel 1', 'oxyHb'], idx[:, :]]
b2 = calc_mean(b, lvl2_trials)

deno = np.sqrt(b2.shape[1])
m2 = b2.apply(np.sum, axis=1)
s2 = b2.apply(np.std, axis=1) / deno

# Visualization 
plt.plot(m2.index, m2, 'r')
plt.fill_between(s2.index, m2-2*s2, m2+2*s2, color='r', alpha=0.2)

plt.plot(m1.index, m1, 'b')
plt.fill_between(s1.index, m1-2*s1, m1+2*s1, color='b', alpha=0.2)
```
